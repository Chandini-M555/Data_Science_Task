To perform Exploratory Data Analysis (EDA) on the Zomato Bangalore dataset and generate

```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset
df=pd.read_excel("zomato_rest_data.xlsx")
df
```

```python
#Identify missing values
df.isnull().sum()
```

**2. Data Cleaning**

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
#drop unnecessary columns
df=df.drop("Unnamed: 0",axis=1)
df=df.drop("url",axis=1)
df=df.drop("address",axis=1)
df=df.drop("phone",axis=1)
df=df.drop("location",axis=1)
df=df.drop("menu_item",axis=1)
df=df.drop("reviews_list",axis=1)
df=df.drop("rest_type",axis=1)
```

```python
#cleaning rate column
#1)removing /5
df["rate"]=df["rate"].str.replace("/5","")
#2)handling new and -
df["rate"]=df["rate"].replace("-",np.nan)
df["rate"]=df["rate"].replace("NEW",np.nan)
#3)changing data type
df["rate"]=df["rate"].astype(float)
#4)handling null values
avg_rate = round(df['rate'].mean())
df['rate'] = df['rate'].fillna(avg_rate)
df["rate"].unique()
```

```python
#changing datatype of approxcost and handling null values
# 1)removing comma
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].str.replace(",","")
#2)changing datatype
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype(float)
#3)handling null values by filling null with mean
mean_cost =round(df["approx_cost(for two people)"].mean())
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].fillna(mean_cost)
df["approx_cost(for two people)"].unique()
```

```python
#rename listed_in(city) column with location
df=df.rename(columns={"listed_in(city)":"location"})
```

```python
#rename listed_in(type) column with rest_type
df=df.rename(columns={"listed_in(type)":"rest_type"})
```

```python
#handling null values in dishliked with mode
mode_dishliked = df["dish_liked"].mode()[0]
df["dish_liked"] = df["dish_liked"].fillna(mode_dishliked)
df["dish_liked"].unique()
```

```python
#handling null values in Cuisine with mode
mode_cuisine = df["cuisines"].mode()[0]
df["cuisines"] = df["cuisines"].fillna(mode_cuisine)
df["cuisines"].unique()
```

```python
# resetting the index
df.reset_index(drop=True,inplace=True)
```

#Task 2: Univariate Analysis

```python
# 1. Top 5 locations with highest number of restaurants using barplot
top5=df["location"].value_counts().head(5)
plt.bar(top5.index,top5,color="r")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.title("Top 5 Locations with Highest Number of Restaurants")
plt.xticks(rotation=90)
plt.show()
```

**Interpretation:**

More no of restuarants are located in BTM

Least no of restuarants are located in Koramangala 6th Block


```python
#Distribution of restaurant ratings either valuecounts or histplot
rating_counts=df["rate"].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.histplot(df["rate"],bins=10,color="r")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

**Interpretation:**

The above graph indicates Normal Distribution and the max rating is in bewteen 3.5 and 4



```python
#Percentage of restaurants offering online orders with pie chart
online_counts=df["online_order"].value_counts()
plt.pie(online_counts,labels=online_counts.index,autopct='%0.5f%%',colors=["red","orange"])
plt.title("Percentage of Restaurants Offering Online Orders")
plt.show()
```

**Interpretation:**

About 58.86% of restuarants offers online orders and 41.13% of restuarants doesnot offer online orders

```python
#4.Distribution of average cost for two w.r.to location
avg_cost_location = df.groupby("location")["approx_cost(for two people)"].mean().sort_values(ascending=False)
sns.histplot(avg_cost_location,color="r")
plt.xticks(rotation=90)
plt.xlabel("cost at that Location")
plt.ylabel("Average Cost for Two")
plt.title("Distribution of Average Cost for Two")

```

**Interpretation:**

The average distribution is maximum i.e, 12 locations have in between 460 to 520

# Task 3: Bivariate Analysis  

```python
#Rating vs Cost
plt.figure(figsize=(10,6))
plt.scatter(x="rate",y="approx_cost(for two people)",data=df,color="r")
plt.xlabel("Rating")
plt.ylabel("Cost")
plt.title("Rating vs Cost")
plt.show()
```

**Conclusion:**

--For the cost between 200 and 2000 has more no of ratings

--If the cost is increasing then no of ratings also increasing





```python
#Rating vs Online Order
Avg_rating=df.groupby("online_order")["rate"].mean()
plt.bar(Avg_rating.index,Avg_rating.values,color="r")
plt.xlabel("Online Order")
plt.ylabel("Average Rating")
plt.title("Average Rating vs Online Order")
plt.show()
```

**Conclusion:**

Even if the restuarant doesnot offer online orders,the average rating is same as the restuarants offering online orders


```python
#Votes vs Rating
plt.figure(figsize=(10,6))
plt.scatter(y="votes",x="rate",data=df,color="r")
plt.ylabel("Votes")
plt.xlabel("Rating")
plt.title("Votes vs Rating")
plt.show()
```

**Conclusion:**

Highest ratings got maximum votes and lowest ratings got less votes
