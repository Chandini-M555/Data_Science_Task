```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/flats_uncleaned.csv",encoding="ISO-8859-1")
df.info()
```

```python
#Removing dupliactes
df.shape
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
#removing null values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#checking evry column
df["property_name"].unique()
```

```python
#checking for extra spaces
df["property_name"] = df["property_name"].str.strip()
df["property_name"].unique()
```

```python
df["areaWithType"].unique() #no issues
```

```python
df["square_feet"].unique()
#here already the column name is square feet...so only numerical value is enough
```

```python
#removing sqft from every record
df["square_feet"]=df["square_feet"].str.strip(" sqft")
df["square_feet"].unique()
```

```python
df["transaction"].unique()
#there are mixing of records here
```

```python
df["transaction"].value_counts() #the actual records should be either New_property or Resale...remaining are errors
```

```python
df["status"].unique()
df["status"].value_counts()
#mixing of different attributes
```

```python
#removing the unwanted column- description
df=df.drop(["description"],axis=1)
df.head()
```

```python
df["floor"].unique()
df["floor"].value_counts()
```

```python
#removing new property and resales from floor column
df["floor"]=df["floor"].replace("New Property","0")
df["floor"]=df["floor"].replace("Resale","0")
df["floor"].value_counts()
```

```python
df["price_per_sqft"].unique()
```

```python
#removing per sqft
df["price_per_sqft"]=df["price_per_sqft"].str.strip(" per sqft")
df["price_per_sqft"].unique()
```

```python
#removing â\x82¹ from the record and comma
df["price_per_sqft"]=df["price_per_sqft"].str.strip("â\x82¹")
df["price_per_sqft"]=df["price_per_sqft"].str.replace(",","")
df["price_per_sqft"].unique()
```

```python
#conversion of data type to int
df["price_per_sqft"]=df["price_per_sqft"].astype(float)
df["price_per_sqft"]=df["price_per_sqft"].astype(int)
df["price_per_sqft"].unique()
```

```python
#remove price column
df=df.drop(["price"],axis=1)
df.head()
```
