```python
#Data cleaning: It is a process of removing data by extracting only wanted data
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/amazon_fires.csv",encoding="ISO-8859-1")
df.info()
```

```python
#1)Renaming the columns
#dict={"old_col":"new_col"}
#syntax: df.rename(columns=dict,inplace=True)
d={"ano":"year",	"mes":"month","estado":"state",	"numero":"no_of_fires",	"encontro":"date"}
df.rename(columns=d,inplace=True)
df
```

```python
#2)Dropping duplicate rows
#syntax: df.drop_duplicates(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
#3)Changing the data present inside columns
#syntax: d={"old_data":"new_data",....}
#df["col"]=df["col"].map(d)
#note: length of dictionary should be equal to unique values
d={"Janeiro":"jan", "Fevereiro":"feb", "Março":"march", "Abril":"april", "Maio":"may", "Junh":"june", "Julho":"july",
       "Agosto":"aug", "Setembro":"sep", "Outubro":"oct", "Novembro":"nov", "Dezembro":"dec"}
df["month"]=df["month"].map(d)
df["month"].unique()
```

```python
#4)Applying inbuilt functions on a column
#syntax:df[col]=df[col].str.fname(args)
df["month"]=df["month"].str.upper()
df["month"]=df["month"].str.capitalize()
df
```

```python
#5)Removing unwanted data
#df["col"]=df["col"].str.strip("unwanted data")
df["no_of_fires"]=df["no_of_fires"].str.strip(" Fires")
df
#here fires after the numeric value is removed
```

```python
#6)Handling missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#7)changing the datatype of columns
#df[col]=df[col].astype(dest_dtype)
df["no_of_fires"]=df["no_of_fires"].astype(float)
df["no_of_fires"]=df["no_of_fires"].astype(int)
df["no_of_fires"].unique()
```

```python
#8)Dropping unwanted columns
#df=df.drop([col1,col2,col3,....],axis=1)
df=df.drop(["state"],axis=1)
df
```

```python
#dropping unwanted rows
#df=df.drop([ind1,ind2])  #axis no need bcoz this is row wise operation
df=df.drop([0,1,2])
df
```

```python
#10)Resetting index
#df.reset_index(inplace=True)
df.reset_index(inplace=True)      #now index column is unwanted
df=df.drop("index",axis=1)
df
```

```python
#Data Cleaning task
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/laptopData.csv",encoding="ISO-8859-1")
df.info()
```

```python
df.isnull().sum()
```

```python
#Handling null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#remove unnamed
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
#drop duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["Company"].unique()
```

```python
df["TypeName"].unique()
```

```python
df["Inches"].unique()
```

```python
#replacing question mark wirh zero and then with avg value
df["Inches"]=df["Inches"].replace("?","0")
df["Inches"]=df["Inches"].astype(float)
avg_inch=round(df["Inches"].mean())
avg_inch
df["Inches"]=df["Inches"].replace(0.0,avg_inch)
df["Inches"].unique()
```

```python
df["Ram"].unique()
```

```python
df["Memory"].unique()
```

```python
#we need to replace question mark with most repeated one
df["Memory"].mode()
df["Memory"]=df["Memory"].replace("?","256GB SSD")
df["Memory"].unique()
```

```python
df["Weight"].unique()
```

```python
#Handling weight
df["Weight"]=df["Weight"].str.strip("kg")
df["Weight"].unique()
df["Weight"]=df["Weight"].replace("?","0")
df["Weight"]=df["Weight"].astype(float)
df["Weight"].mean()
df["Weight"]=df["Weight"].replace(0.0,2.0785)
df["Weight"].unique()
```

```python
df.reset_index(inplace=True)
df=df.drop("index",axis=1)
df
```
