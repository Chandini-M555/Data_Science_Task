```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Taj1920/Data_Analysis/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
df["Subscribers"].unique()
```

```python
df["Ranking"]=df["Ranking"].fillna(101)
df["Ranking"].unique()
```

```python
df["Username"]=df["Username"].fillna("Unknown")
df["Username"].unique()
```

```python
df["Subscribers"]=df["Subscribers"].str.strip("M")
```

```python
df["Subscribers"]=df["Subscribers"].astype(float)
df["Subscribers"].unique()
```

```python
df["Subscribers"]=df["Subscribers"].fillna(df["Subscribers"].mean())
df["Subscribers"].unique()
```

```python
df["Uploads"].unique()
```

```python
df["Uploads"]=df["Uploads"].fillna("0")
df["Uploads"].unique()
```

```python
df["Views"].unique()
```

```python
df["Views"]=df["Views"].fillna("0")
df["Views"].unique()
```

```python
df["Country"].unique()
df["Country"]=df["Country"].fillna("Unknown")
```

```python
df["Country"].unique()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["Views"]=df["Views"].str.replace(",","")
df.head()
```

```python
df["Uploads"]=df["Uploads"].str.replace(",","")
df.head()
```

```python
df
```

```python
plt.plot(df["Subscribers"],df["Ranking"])
plt.xlabel("Sub")
plt.ylabel("Rank")
```

```python
plt.bar(df["Country"],df["Subscribers"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Subscribers")
plt.show()
```

US has most subscribers

```python
plt.bar(df["Country"],df["Uploads"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Uploads")
plt.show()
```
