```python
#creating and reshaping
#creating arrays
import numpy as np
arr1=np.random.randint(0,20,10)
arr2=np.random.randint(5,20,9)
arr2=arr2.reshape(3,3)
print(arr1)
print(arr2)
```

```python
#array dimensions
arr1.ndim
```

```python
arr2.ndim
```

```python
#reshaping 1-d to 2-d
arr1.reshape(2,5)
```

```python
#reshaping 2-d again to 1-d
arr1.reshape(-1)
```

```python
#indexing
print(arr2[1,2]) -->element at 2nd row and third column
```

```python
#slicing
print(arr2[0:2,1:3])
print(arr1[0:4])
arr1[::-1]
```

```python
#array arithmetic
arr3=np.arange(1,11)
print(arr3)
print(arr1+arr3)
print(arr1*arr3)
print(arr3-5)
```

```python
#random  numbers
arr4=np.random.rand(5)
arr5=np.random.randn(5)
print(arr4)
print(arr5)
```
